```python
df=pd.DataFrame({"height":np.random.randint(10,1000,10),"Date":date1})
plt.plot(df['Date'],df['height'])

# help(plt.colormaps)
```

```python
df.index=df.Date
#df.drop(['Date'],axis=1,inplace=True)
df.plot()
plt.plot(df['Date'])
plt.plot(df['height'])
```
